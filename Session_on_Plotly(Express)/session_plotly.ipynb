```python
trace1 =go.Scatter(x = temp_df['lifeExp'],y = temp_df['gdpPercap'])
data=[trace1]
layout = go.Layout(title='life ex graph',xaxis={'life'},yaxis={'gdp'})
fig = go.Figure(data,layout)

fig.show()
```

```python
import plotly.graph_objects as go

# Assuming temp_df is already defined
trace1 = go.Scatter(x=temp_df['lifeExp'], y=temp_df['gdpPercap'], mode='markers')

data = [trace1]
layout = go.Layout(
    title='Life Expectancy vs GDP',
    xaxis={'title': 'Life Expectancy'},
    yaxis={'title': 'GDP per Capita'}
)

fig = go.Figure(data=data, layout=layout)
fig.show()
```
